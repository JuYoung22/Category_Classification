```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time

options = webdriver.ChromeOptions()
options.add_argument('--headless')  # 브라우저를 표시하지 않고 실행하려면 이 옵션을 추가합니다.
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# 코랩에서 크롬 드라이버 실행
driver =  webdriver.Chrome(options=options)

def webtoonData(inputWebtoonLink):
    driver = webdriver.Chrome(options=options)

    url = inputWebtoonLink
    driver.get(url)

    a_tags = driver.find_elements(By.CSS_SELECTOR, '.relative.flex.items-center.justify-center a')
    not_link = []  # not_link 변수 초기화

    if len(a_tags) > 1:
        second_link = a_tags[1]
        second_link.click()

        time.sleep(1)

        try:
            # 원하는 요소를 CSS 선택자로 찾기
            title_element = driver.find_element(By.CSS_SELECTOR, 'span.font-large3-bold.mb-3pxr.text-ellipsis.break-all.text-el-70.line-clamp-2')
            summary_element = driver.find_element(By.CSS_SELECTOR, 'span.font-small1.mb-8pxr.block.whitespace-pre-wrap.break-words.text-el-70')
            # div 태그 내에서 a 태그를 찾아서 data-t-obj 속성 값을 가져오기
            category_tag_elements = driver.find_elements(By.CSS_SELECTOR, 'div.flex.w-full.flex-wrap.px-18pxr.pb-10pxr a.mb-8pxr.mr-8pxr')

            # 데이터 프레임 생성을 위한 빈 리스트 초기화
            titles = []
            summaries = []
            category_tags = []


            # 요소에서 텍스트 추출 및 리스트에 추가
            if title_element:
                text_title = title_element.text
                text_summary = summary_element.text
                text_category_tag = []

                for element in category_tag_elements:
                    data_t_obj = element.get_attribute('data-t-obj')

                    if 'click' in data_t_obj:
                        copy_info = data_t_obj.split('"copy":')[1].split('}')[0]
                        text_category_tag.append(copy_info.strip('"'))

                category_tags_combined = ', '.join(text_category_tag)
                text_summary = text_summary.replace('\n', "")

                # 리스트에 추가
                titles.append(text_title)
                summaries.append(text_summary)
                category_tags.append(category_tags_combined)

            else:
                # print(f"두 번째 링크에 대한 정보가 존재하지 않습니다: {url}")
                driver.quit()
                return None, [url], 1  # 수정: not_link을 리스트로 반환하고, current_count도 함께 반환

        except NoSuchElementException as e:
            # print(f"두 번째 링크에 대한 정보를 찾을 수 없습니다: {e}")
            driver.quit()
            return None, [url], 1  # 수정: not_link을 리스트로 반환하고, current_count도 함께 반환

    else:
        # print(f"두 번째 링크가 존재하지 않습니다: {url}")
        not_link.append(url)
        driver.quit()
        return None, not_link, 1  # 수정: not_link을 리스트로 반환하고, current_count도 함께 반환

    driver.quit()

    # 데이터 프레임 생성
    df = pd.DataFrame({
        'Title': titles,
        'Summary': summaries,
        'Category_Tag': category_tags
    })

    return df, not_link, 1  # 수정: not_link을 리스트로 반환하고, current_count도 함께 반환

# 예시로 200번 수집할 때까지 호출
category_list = q[1300:]
result_dfs = []
result_not_links = []

for index, link in enumerate(category_list, start=1):
    result_df, not_links, current_count = webtoonData(link)

    if result_df is not None:
        result_dfs.append(result_df)
    else:
        result_not_links.extend([(link, count) for count in not_links])

    # 100번마다 현재 수집 정도 출력
    if index % 5 == 0:
        print(f"현재 수집 정도: {index}")

# 최종 결과 데이터프레임 생성
final_df = pd.concat(result_dfs, ignore_index=True)

print(f"최종 수집 완료: {len(final_df)}")
```

```python
# 토큰화된 데이터에 웹툰 제목도 추가
tokenized_data = []
for index, text in enumerate(webtoon_text_data):
    tokens = text.split()
    tokens.extend(tokens[0])  # 웹툰 제목 추가
    tokenized_data.append(tokens)
```

```python
# Word2Vec 모델 학습
model = Word2Vec(sentences=tokenized_data, vector_size=5000, window=3, min_count=1, workers=4)

# 완성된 임베딩 매트릭스의 크기 확인
embedding_size = model.wv.vectors.shape
print("Embedding Matrix Size:", embedding_size)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# 유사도 계산을 위한 함수: 코사인 유사도 사용.
def calculate_similarity(title1, title2):
    vec1 = model.wv[title1]
    vec2 = model.wv[title2.replace(" ", "")]
    similarity = cosine_similarity([vec1], [vec2])[0][0]

    # 백분율
    similarity_percentage = (similarity + 1) * 50
    return similarity_percentage

# 사용자 입력 받기
user_input = input("유사한 웹툰을 찾을 키워드를 입력하세요: ")

# 입력을 기반으로 유사한 웹툰 추천
target_webtoon = user_input
recommended_webtoons = train_data.copy()  # 복사본 생성

# 'Similarity' 컬럼 추가하지 않고 바로 출력
similarities = []

print(f'\n{target_webtoon}과(와) 유사한 웹툰 추천:\n')
top_n = 10  # 원하는 상위 N개의 웹툰을 출력하도록 설정
count = 0

for index, row in recommended_webtoons.iterrows():
    similarity = calculate_similarity(target_webtoon, row['Title'])
    similarities.append((row['Title'], similarity))

# 유사도가 높은 순으로 정렬하여 상위 N개 출력
similarities = sorted(similarities, key=lambda x: x[1], reverse=True)[:top_n]
for title, similarity in similarities:
    count += 1
    print(f"{count}. {title} : 유사도: {similarity:.2f}%")
```

### 한국어 KoBERT

```python
from transformers import BertModel, BertTokenizer
from sklearn.metrics.pairwise import cosine_similarity
import torch
```

```python
# koBERT 모델과 토크나이저 불러오기
model_name = 'kykim/bert-kor-base'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# 입력 텍스트를 koBERT의 임베딩 벡터로 변환하는 함수
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True)
    outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1)
```

```python

# 사용자 입력 받기
user_input = input("추천을 원하는 웹툰을 입력하세요: ")

# 입력 텍스트와 각 웹툰 간의 유사도 계산
user_embedding = get_embedding(user_input)
webtoon_embeddings = [get_embedding(text) for text in webtoon_text_data]
similarities = cosine_similarity(user_embedding, webtoon_embeddings).flatten()

# 유사도가 높은 순으로 정렬하여 상위 5개 출력
sorted_similarities = sorted(enumerate(similarities), key=lambda x: x[1], reverse=True)[:5]


print(f'\n"{user_input}"과(와) 유사한 웹툰 추천:\n')

for i, (index, similarity) in enumerate(sorted_similarities, start=1):
    title = f"웹툰{index + 1}"  # 실제 데이터에 맞게 수정 필요
    print(f"{i}. {title} : 유사도: {similarity:.2f}")
```
